Exercício Prático:

A nova missão de vocês agora é **prever o preço de casas.**

Faça todos os passos de treinar os modelos em Random Forest, AdaBoost e Gradient Boosting.

E tira para cada algoritmo a feature importance.

```python
# Importando as bibliotecas

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.datasets import make_moons
from sklearn import metrics
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
```

```python
# Importando os dados

casa = pd.read_excel("/content/house (1).xlsx")
casa.head()
```

```python
# Imformação da base

casa.info()
```

```python
# eliminado dados faltantes

casa = casa.dropna()
```

```python
# descrição estatistica da base

casa.describe()
```

```python
# transformando objetos em dummies

casa = pd.get_dummies(casa)
casa.head()
```

```python
# Agora vamos transformar a base x e y com pandas

labels = np.array(casa['SalePrice'])
casa = casa.drop('SalePrice', axis=1)
casa_list = list(casa.columns)
features = np.array(casa)
```

```python
# separando base treino teste

train_casa, test_casa, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state=42)
```

```python
# treinando modelo

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(train_casa, train_labels)
```

```python
# Performa-se do modelo para Random Forest


predictions_rf = rf.predict(test_casa)

erros_rf = abs(predictions_rf - test_labels)

r_sq = rf.score(casa, labels)
print("R^2:", r_sq)
print("MAE:", metrics.mean_absolute_error(test_labels,predictions_rf ))
print("MSE:", metrics.mean_squared_error(test_labels,predictions_rf ))
```

```python
#  Performa-se do modelo para AdaBoost

ada = AdaBoostRegressor(n_estimators=1000, random_state=42)

ada.fit(train_casa, train_labels);

predictions_ada = ada.predict(test_casa)

erros_ada = abs(predictions_ada - test_labels)

r_sq = ada.score(casa, labels)
print("R^2:", r_sq)
print("MAE:", metrics.mean_absolute_error(test_labels,predictions_ada ))
print("MSE:", metrics.mean_squared_error(test_labels,predictions_ada ))
```

```python
# Performa-se do modelo para GradientBoosting

gbr = GradientBoostingRegressor(n_estimators=1000, random_state=42)
gbr.fit(train_casa, train_labels);

predictions_gbr = gbr.predict(test_casa)

erros_gbr = abs(predictions_gbr - test_labels)

r_sq = gbr.score(casa, labels)
print("R^2:", r_sq)
print("MAE:", metrics.mean_absolute_error(test_labels,predictions_gbr ))
print("MSE:", metrics.mean_squared_error(test_labels,predictions_gbr ))
```

Essa métrica ficou ainda melhor que o AdaBoosting, portanto esse é o nosso melhor modelo
